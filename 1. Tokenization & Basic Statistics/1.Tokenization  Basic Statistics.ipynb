#Exercise 2

#### SHERLOCK

**Concordance**

```python
tokenized_sent = word_tokenize(sherlock)
text = Text(tokenized_sent)

words = ["Sherlock", "extreme"]
for word in words:
  print(f"\nConcordance of word {word} is:")
  text.concordance(word)
```

**Common contexts**

```python
words = ["gathering", "extreme"]
text.common_contexts(words, num=20)
```

**Collocations**

```python
finder = BigramCollocationFinder.from_words(text)

top_bigrams = finder.nbest(BigramAssocMeasures.pmi, 100)
print(top_bigrams)
```

```python
all_words = [word for fileid in inaugural.fileids() for word in inaugural.words(fileids=fileid)]
```

### 4.2. Calculate the frequency distribution of every word in the vocabulary


```python
from collections import Counter
counts = Counter(all_words)
print(counts)
```

### 4.3. Plot the frequency distribu8on of the top 30 frequently used words


```python
top30 = counts.most_common(30)
top30 = dict(top30)
print(top30)
```

```python
plt.figure(figsize=(15,6))
plt.bar(top30.keys(), top30.values(), width=0.5)
plt.show()
```

### 4.4. Plot the cumulative frequency distribution of the same words

```python
top30_sorted_dict = dict(sorted(top30.items(), key=lambda item: item[1], reverse=True))

cumulative = []
total_freq = 0
for freq in top30_sorted_dict.values():
    total_freq += freq
    cumulative.append(total_freq)

plt.figure(figsize=(15, 6))
plt.plot(range(1, len(cumulative) + 1), cumulative, marker='o', linestyle='-')
plt.title('Cumulative Frequency Distribution')
plt.xlabel('Rank')
plt.ylabel('Cumulative Frequency')
plt.grid(True)
plt.show()
```

### 4.5. Select another text, plot its frequency distribution, and show the first 50 hapaxes
